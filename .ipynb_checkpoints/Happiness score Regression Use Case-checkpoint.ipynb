#### 3.1 Предобработка данных 

```python
# Приводим названия полей к одному единому виду в датасете за 2015 год
for col in df_2015.columns:
    col_corrected = col.strip().replace(' ','_').replace('(','').replace(')','').lower()
    df_2015.rename(columns={col:col_corrected}, inplace = True)
# Приводим названия полей к одному единому виду в датасете за 2016 год   
for col in df_2016.columns:
    col_corrected = col.strip().replace(' ','_').replace('(','').replace(')','').lower()
    df_2016.rename(columns={col:col_corrected}, inplace = True)
# Приводим названия полей к одному единому виду в датасете за 2017 год   
for col in df_2017.columns:
    col_corrected = col.strip().replace('..',' ').replace('.',' ').replace('(','').replace(')','').replace(' ','_').lower()
    
    if col_corrected in ['health_life_expectancy_', 'trust_government_corruption_', 'economy_gdp_per_capita_']:
        col_corrected = col_corrected[:len(col_corrected)-1]
        
    df_2017.rename(columns={col:col_corrected}, inplace = True)
# Приводим названия полей к одному единому виду в датасете за 2018 год      
for col in df_2018.columns:
    col_corrected = col.strip().replace(' ','_').replace('(','').replace(')','').lower()
    df_2018.rename(columns={col:col_corrected}, inplace = True)
# Приводим названия полей к одному единому виду в датасете за 2019 год  
for col in df_2019.columns:
    col_corrected = col.strip().replace(' ','_').replace('(','').replace(')','').lower()
    df_2019.rename(columns={col:col_corrected}, inplace = True)
    
# Переименовываем некоторые колонки в датасете за 2019 год, чтобы они совпадали с названиями в остальных датасетах
df_2019.rename(columns = {'overall_rank': 'happiness_rank',
                          'score': 'happiness_score',
                          'gdp_per_capita': 'economy_gdp_per_capita',
                          'country_or_region': 'country',
                          'freedom_to_make_life_choices': 'freedom',
                          'perceptions_of_corruption': 'trust_government_corruption',
                          'social_support':'family',
                          'healthy_life_expectancy':'health_life_expectancy'
                         }, inplace= True)
# Аналогично и в датасете в 2018 году
df_2018.rename(columns = {'overall_rank': 'happiness_rank',
                          'score': 'happiness_score',                         
                          'gdp_per_capita': 'economy_gdp_per_capita',
                          'country_or_region': 'country',
                          'freedom_to_make_life_choices': 'freedom',
                          'perceptions_of_corruption': 'trust_government_corruption',
                          'social_support':'family',
                          'healthy_life_expectancy':'health_life_expectancy'
                         }, inplace= True)
```

## 4. Создаем новые признаки

```python
# Объединяем все наши датафреймы в один с новым признаком year 
df_2015['year'] = 2015
df_2016['year'] = 2016
df_2017['year'] = 2017
df_2018['year'] = 2018
df_2019['year'] = 2019
df_total = pd.concat((df_2015,df_2016,df_2017,df_2018, df_2019), axis = 0)
# Проверяем кол-во строк исходных датафреймов с кол-вом строк единого датафрейма
print(df_total.shape[0] == (df_2015.shape[0]+df_2016.shape[0]+df_2017.shape[0]+df_2018.shape[0]+df_2019.shape[0]))

# Удаляем признаки happinnes_rank с очень негативно влияющей на наш таргет и dystopia_residual, где много NaN значений
df_total.drop(columns = ['happiness_rank', 'dystopia_residual'], inplace = True)
```

```python
# С помощью Mean Target Encoding кодируем признак country и добавляем немного шума чтобы избежать переобучения
mean_score_by_country = df_total.groupby('country').happiness_score.mean() 

for country in mean_score_by_country.index:
    mean_score_by_country[country] = mean_score_by_country[country] + np.random.uniform(0,1) * 2   
    
mean_score_by_country = mean_score_by_country.reset_index()
mean_score_by_country.rename(columns = {'happiness_score':'country_mte'},inplace = True)
df_total = df_total.merge(mean_score_by_country)
df_total.drop(columns = 'country', inplace = True)
df_total_2 = df_total.copy()
# С помощью One Hot Encoding'а кодируем признак region
region_ohe = pd.get_dummies(df_total_2['region'],drop_first = True)
df_total_2 = df_total.merge(region_ohe, left_index = True, right_index=True)
df_total_2.drop(columns = 'region', inplace = True)
```

```python
# Cравниваем кол-во столбцов до feature engineering'а и после
print(df_total.shape)
print(df_total_2.shape)
```

## 5.Отбираем признаки 

```python
# Сортируем наш объединенный датафрейм в один и сортируем его по годам
df_total_2 = df_total_2.sort_values('year')
# Разделяем наши признаки и таргет в разные сущности
X = df_total_2.drop(columns = 'happiness_score')
y = df_total_2.happiness_score
```

```python
# Проверяем дисперсию у наших признаков, в случае если она очень маленькая, то выбрасываем такую фичу
variance = VarianceThreshold(threshold = 0.1)
variance.fit(X)
keep_columns_list = variance.get_feature_names_out()
X_ = X[keep_columns_list] 
```

```python
# Финальный лист наших признаков
keep_columns_list
```

```python
# Сравниваем кол-во признаков до feature selection'а и после.
print(df_total_2.shape)
print(X_.shape)
```

## 6. Подбор и обучение моделей.

```python
# Разбиваем наш датафрейм с признаками X и таргет y на train, test выборки с shuffle = false, так как у нас timeseries датафрейм
X_train, X_test, y_train, y_test = train_test_split(X_, y, test_size=0.2, shuffle = False, random_state = 42)
# Бейзлайн модель в виде обычной Лин рега
# Тестируем перфоманс модели с учетом фичей и feature engineering'а 
np.random.seed = 42
pipe_LR = Pipeline([ ('scaler', StandardScaler()), ('LR', LinearRegression())])
pipe_LR.fit(X_train, y_train)
y_preds = pipe_LR.predict(X_train)
y_HOLDOUT = pipe_LR.predict(X_test)

print(f"LR MSE train score: {round(mean_squared_error(y_train, y_preds),3)}")
print(f"LR MSE test score: {round(mean_squared_error(y_test, y_HOLDOUT),3)}")
```

```python
# Следующая модель - это дерево решений
from sklearn.tree import DecisionTreeRegressor

pipe_DTR = Pipeline([('DTR', DecisionTreeRegressor(random_state = 42))])
pipe_DTR.fit(X_train, y_train)
y_preds = pipe_DTR.predict(X_train)
y_HOLDOUT = pipe_DTR.predict(X_test)

print(f"Tree MSE train score: {round(mean_squared_error(y_train, y_preds),3)}")
print(f"Tree MSE test score: {round(mean_squared_error(y_test, y_HOLDOUT),3)}")
```

Можем видеть, что дерево решений без настройки гиперпараметров переобучилась.

```python
# Попробуем бустинг XGB в качестве третьей модели
pipe_XGB = Pipeline([ ('scaler', StandardScaler()), ('XGB', xgb.XGBRegressor(random_state = 42))])
pipe_XGB.fit(X_train, y_train)
y_preds = pipe_XGB.predict(X_train)
y_HOLDOUT = pipe_XGB.predict(X_test)
print(f"XGB MSE train score: {round(mean_squared_error(y_train, y_preds),3)}")
print(f"XGB MSE test score: {round(mean_squared_error(y_test, y_HOLDOUT),3)}")
```

## 7.Ищем наилучшие гиперпараметры наших моделей
